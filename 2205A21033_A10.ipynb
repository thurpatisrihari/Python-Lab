```python
import numpy as np
marks=np.array([[[101,35,98,43,65,78,96],
                 [102,87,45,98,65,34,78],
                 [103,45,26,78,45,67,34],
                 [104,45,56,67,78,89,90],
                 [105,23,34,45,56,76,87]],
                [[101,45,89,43,56,87,69],
                 [102,78,54,89,56,43,87],
                 [103,54,62,87,54,76,43],
                 [104,35,65,76,87,98,9],
                 [105,32,43,54,65,67,78]]])
students=marks.shape[1]
sem=marks.shape[0]


Roll_no=int(input("enter the Roll_no:"))
semester=int(input("enter the Semester:"))
for i in range(sem):
  if(semester==i+1):
    break

index_sem=i
for j in range(students):
  if(Roll_no==marks[index_sem,j,0]):
    break

index_students=j
percentages=np.sum(marks[index_sem,index_students,1:marks.shape[2]])*100/600
count=0
failed_sub=[]
for i in range(1,marks.shape[2]):
  if(marks[index_sem,index_students][i]<40):
    count=count+1
    failed_sub.append(i)
print("===========================")
print(" SR UNIVERSITY ")
print(" score card ")
print("===========================")
print("Roll_no:",Roll_no)
print("semester:",semester)
print("subject 1:",marks[index_sem,index_students,1])
print("subject 2:",marks[index_sem,index_students,2])
print("subject 3:",marks[index_sem,index_students,3])
print("subject 4:",marks[index_sem,index_students,4])
print("subject 5:",marks[index_sem,index_students,5])
print("subject 6:",marks[index_sem,index_students,6])
print("==============================")
print(" Results ")
print("==============================")
if(count>0):
  print("faiil")
  print("failed subject:",)
  for i in range(len(failed_sub)):
    if(failed_sub[i]==1):
      print("subject 1")
    if(failed_sub[i]==2):
      print("subject 2")
    if(failed_sub[i]==3):
      print("subject 3")
    if(failed_sub[i]==4):
      print("subject 4")
    if(failed_sub[i]==5):
      print("subject 5")
    if(failed_sub[i]==6):
      print("subject 6")
else:
  print("pass")
  print("percentage:",np.round(percentages))
print("=====================================")
```
